```python
'''
Load and Join Data Example
'''

import pandas as pd
import os
```

```python
label_data = pd.read_csv(os.path.join(os.getcwd(), 'crowd_train.csv'))
```

```python
subreddit_data = pd.read_csv(os.path.join(os.getcwd(), 'task_B_train.posts.csv'))
```

```python
text_data = pd.read_csv(os.path.join(os.getcwd(), 'shared_task_posts.csv'))
```

```python
sub_label_data = pd.merge(subreddit_data, label_data, how = 'left', on = 'user_id')
_data_ = pd.merge(sub_label_data, text_data, on = ['post_id', 'user_id'])
```
